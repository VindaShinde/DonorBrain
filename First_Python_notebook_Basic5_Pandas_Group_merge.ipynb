```python
data = pd.read_csv('report.csv')
```

```python
print(data.describe)
data.head()
```

```python
grpByJurisdic = data.groupby('agency_jurisdiction')
print(grpByJurisdic.groups)
len(grpByJurisdic.groups)
```

```python
grpByMonths = data.groupby('months_reported')
monthsReports = grpByMonths.size()
monthsReports.sort_values()
```

```python
crimesGroups = data.groupby(['agency_jurisdiction','months_reported'])
crimes = crimesGroups.size()
crimes.sort_index()
```
